```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("laptops.csv")
df
```

```python
df.drop(["id","model","comments"],axis="columns",inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df = df.fillna({"graphic_card_size":0,"weight":df.weight.mean(),"clock_speed":df.clock_speed.mean()})
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.brand.value_counts()
```

```python
df.brand.replace({"Asus ":"Asus","HP ":"HP"},inplace=True)
df
```

```python
df.brand.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
brand_encoder = LabelEncoder()
df.brand = brand_encoder.fit_transform(df.brand)
brand_encoder.classes_
```

```python
df.ram.dtype
```

```python
hdtype_encoder = LabelEncoder()
df.hd_type = hdtype_encoder.fit_transform(df.hd_type)
df
```

```python
df.screen_size.dtype
```

```python
df1 = df[df.processor_brand=="intel"]
```

```python
df1.drop("processor_brand",axis="columns",inplace=True)
df1
```

```python
df1.processor_model.value_counts()
```

```python
def pro(x):
    x=x.lower()
    y=x.split(" ")[0]
    if y.startswith("i"):
        return y
    else:
        return "other"
```

```python
df1.processor_model = df1.processor_model.apply(pro)
```

```python
df1
```

```python
processor_model_encoder = LabelEncoder()
df1.processor_model = processor_model_encoder.fit_transform(df1.processor_model)
processor_model_encoder.classes_
```

```python
df1.clock_speed.dtype
```

```python
df1.graphic_card_brand.value_counts()
```

```python
graphic_brand_encoder = LabelEncoder()
df1.graphic_card_brand = graphic_brand_encoder.fit_transform(df1.graphic_card_brand)
graphic_brand_encoder.classes_
```

```python
df2=df1[df1.os!="windows,linux"]
```

```python
os_encoder = LabelEncoder()
df2.os = os_encoder.fit_transform(df2.os)
os_encoder.classes_
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df2.drop("price",axis="columns"),df2.price,test_size=0.1)
```

```python
from sklearn.linear_model import LinearRegression
l = LinearRegression()
l.fit(x_train,y_train)
```

```python
import joblib
joblib.dump(l,"model_trial")
l.score(x_test,y_test) , l.score(x_train,y_train)
```
